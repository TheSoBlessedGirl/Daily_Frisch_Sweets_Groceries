```python
import requests
from bs4 import BeautifulSoup

import csv

#Fetch the url

url = "https://en.daily-frisch.de/usa-candy-1"
response1 = requests.get(url)

#Parse the html content using BeautifulSoup

soup1 = BeautifulSoup(response1.content, 'html.parser')

#Extract the products information

print(soup1.find_all('li', {'data-hook':'product-list-grid-item'}))
try:
    all_items =  soup1.find_all('li', {'data-hook':'product-list-grid-item'})
except:
    all_items = []

for each in all_items :
    response = requests.get(each.a['href'])
    soup = BeautifulSoup(response.content, 'html.parser')
    try:
        product_name = soup.find('h1',class_='_2qrJF igTU-').text
    except:
        product_name = ''
    product_price = soup.find('span', {'data-hook':'formatted-primary-price'}).text
    content_quantity = soup.find('pre',class_='_28cEs').find('p').text
    content_description = soup.find('pre',class_='_28cEs').find_all('p')[-1].text
    #product_description = soup.find('div',class_='_300vE').img['src']
    product_quantity = soup.find('input',{'data-hook':'number-input-spinner-input'})['value']
    print(product_name,product_price,content_quantity)
```
